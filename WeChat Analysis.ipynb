```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '~/Downloads/wechat_data_analysis/WeChat_Article_Tracker1.csv'
```

```python
wechat = pd.read_csv(path,
                    header = 1,index_col = 'Date')
wechat.head()
```

```python
page_view = wechat['Page View']
share = wechat['Share']
like = wechat['Like']
ax = sns.distplot(page_view, rug=True)
plt.title('Estimated Density of Page View Counts in WeChat Articles')
```

```python
wechat.plot.scatter(x = 'Page View', y = 'Like')
plt.title("Page View Counts vs Like Counts in WeChat Articles")
wechat.plot.scatter(x = 'Page View', y = 'Share')
plt.title("Page View Counts vs Share Counts in WeChat Articles")
```

```python
corr1float = np.corrcoef(page_view, like)[0][1]
corr1 = "%.2f" % corr1float
print("The correlation between page view abnd like is " + corr1)
corr2float = np.corrcoef(page_view, share)[0][1]
corr2 = "%.2f" % corr2float
print("The correlation between page view abnd like is " + corr2)
```

```python
sns.distplot(share, rug=True)
plt.title('Estimated Density of Share Counts in WeChat Articles')
```

```python
sns.distplot(like, rug=True)
plt.title('Estimated Density of Like Counts in WeChat Articles')
```

```python
wechat['Title Length'] = wechat['Article Title'].str.len()
sns.regplot(x="Title Length", y="Page View", data=wechat)
plt.title("Page View Counts vs Title Length in WeChat Articles")
```
